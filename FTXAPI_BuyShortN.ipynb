# 整合

```python
import time, json
import hmac
import math
import requests
import pandas as pd
from time import sleep
import FTX_Class
from requests import Request
from operator import itemgetter

ts = int(time.time() * 1000)
ts2 = int(time.time())-3601
apiurl = 'https://ftx.com/api'

YOUR_API_KEY = ''
YOUR_API_SECRET = ''

#查看資金費率
endpoint = '/funding_rates?start_time='+str(ts2)
request = Request('GET', apiurl+endpoint)
prepared = request.prepare()
signature_payload = f'{ts2}{prepared.method}{prepared.path_url}'.encode()
signature = hmac.new(YOUR_API_SECRET.encode(), signature_payload, 'sha256').hexdigest()
prepared.headers['FTX-KEY'] = YOUR_API_KEY
#有子帳戶才要下面這行
#request.headers['FTX-SUBACCOUNT'] = 'API_TEST'
prepared.headers['FTX-SIGN'] = signature
prepared.headers['FTX-TS'] = str(ts2)  
s = requests.Session()
rs=s.send(prepared)

RT = rs.json()['result']
#依照資金費率大到小排序
ST = sorted(RT,key=itemgetter('rate'),reverse=True)
df = pd.DataFrame(ST)
#只取幣的名字
df['future'] = df['future'].str.replace("-PERP","")


#列出現貨市場所有幣
endpoint = '/markets'
request = Request('GET', apiurl+endpoint)
prepared = request.prepare()
signature_payload = f'{ts}{prepared.method}{prepared.path_url}'.encode()
signature = hmac.new(YOUR_API_SECRET.encode(), signature_payload, 'sha256').hexdigest()
prepared.headers['FTX-KEY'] = YOUR_API_KEY
prepared.headers['FTX-SIGN'] = signature
prepared.headers['FTX-TS'] = str(ts)  

s = requests.Session()
rs=s.send(prepared)
RT = rs.json()['result']
#pd.options.display.max_rows=None
dfM = pd.DataFrame(RT)
#模糊搜尋包含USD 不包含USDT的目標
dfM2 = dfM.loc[dfM['name'].str.contains('USD') & ~dfM['name'].str.contains('USDT')]
#比對現貨及合約都有的幣放入陣列
Target = []
for i in range(411):
    try:
        if len(dfM2.loc[dfM2['name'].str.contains(df['future'][i])]) == 1:
            Target.append(df['future'][i])
    except:
        pass
    
NTarget=""
NTarget=Target[0]
ts = int(time.time() * 1000)
apiurl = 'https://ftx.com/api'
endpoint = '/markets/'+NTarget+'/USD'
request = Request('GET', apiurl+endpoint)
prepared = request.prepare()
signature_payload = f'{ts}{prepared.method}{prepared.path_url}'.encode()
signature = hmac.new(YOUR_API_SECRET.encode(), signature_payload, 'sha256').hexdigest()
prepared.headers['FTX-KEY'] = YOUR_API_KEY
prepared.headers['FTX-SIGN'] = signature
prepared.headers['FTX-TS'] = str(ts)  


s = requests.Session()
rs=s.send(prepared)
MarketPrice = rs.json()['result']['price']
MarketPrice

#查看錢包
endpoint = '/wallet/balances'
request = Request('GET', apiurl+endpoint)
prepared = request.prepare()
signature_payload = f'{ts}{prepared.method}{prepared.path_url}'.encode()
signature = hmac.new(YOUR_API_SECRET.encode(), signature_payload, 'sha256').hexdigest()
prepared.headers['FTX-KEY'] = YOUR_API_KEY
prepared.headers['FTX-SIGN'] = signature
prepared.headers['FTX-TS'] = str(ts)  
s = requests.Session()
rs=s.send(prepared)
dfWallet = pd.DataFrame(rs.json()['result'])
dfWallet
dfWallet2 = dfWallet.loc[dfWallet['coin'].str.contains('USD') & ~dfWallet['coin'].str.contains('USDT')]
USDinWallet = int(dfWallet2['total'])/2
#算出買多少現貨取小數3位
Buy = int(USDinWallet) / MarketPrice
Buy = math.floor(Buy * 1000) / 1000.0
#回推用多少U開空
#Short = math.floor((MarketPrice*Buy) * 1000) / 1000.0
print('幣種: '+NTarget+'\nPrice: '+str(MarketPrice)+'\n資金費率: '+str(ST[0]['rate'])+'\n錢包裡的資金: '+str(int(dfWallet2['total']))+'\nBuy: '+str(Buy))
```

```python
#紀錄本次買了多少數量
CurrentBuy = Buy
CurrentBuy
```

```python
#參數分別為: 交易對 / 買或賣 / 要掛的價格 / 數量
def BuyCoin():
    c = FTX_Class.FtxClient(api_key=YOUR_API_KEY, api_secret=YOUR_API_SECRET)
    try:
        r = c.place_order(NTarget+"/USD", "buy","0",str(Buy))
        print(r)
    except Exception as e:
        print(f'Error making order request: {e}')

def SellCoin():
    c = FTX_Class.FtxClient(api_key=YOUR_API_KEY, api_secret=YOUR_API_SECRET)
    try:
        r = c.place_order(NTarget+"/USD", "sell","0",str(CurrentBuy))
        print(r)
    except Exception as e:
        print(f'Error making order request: {e}')
def BuyPERP():
    c = FTX_Class.FtxClient(api_key=YOUR_API_KEY, api_secret=YOUR_API_SECRET)
    try:
        r = c.place_order2(NTarget+"-PERP", "buy","0",str(Buy))
        print(r)
    except Exception as e:
        print(f'Error making order request: {e}')

def SellPERP():
    c = FTX_Class.FtxClient(api_key=YOUR_API_KEY, api_secret=YOUR_API_SECRET)
    try:
        r = c.place_order(NTarget+"-PERP", "sell","0",str(CurrentBuy))
        print(r)
    except Exception as e:
        print(f'Error making order request: {e}')

def CoinProtect():
    c = FTX_Class.FtxClient(api_key=YOUR_API_KEY, api_secret=YOUR_API_SECRET)
    try:
        r = c.place_order_Protect(NTarget+"/USD", "sell",str(MarketPrice*2),str(CurrentBuy))
        print(r)
    except Exception as e:
        print(f'Error making order request: {e}')
        
def Startarbitrage():
    BuyCoin()
    time.sleep(0.1)
    SellPERP()
    #time.sleep(1)
    #CoinProtect()
def Closearbitrage():
    SellCoin()
    time.sleep(0.1)
    BuyPERP()
    
print("OK-----"+time.strftime('%Y-%m-%d %H:%M:%S', time.localtime()))
```
